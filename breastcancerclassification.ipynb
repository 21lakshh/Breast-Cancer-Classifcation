```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
import pickle
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
breast_cancer_dataframe = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# Malignant tumors are considered cancerous as they grow rapidly and can travel to other parts of the body
# Benign tumors are non cancerous and their growth is very slow and do not invade other parts of the body
breast_cancer_dataframe['label'] = breast_cancer_dataset.target
```

```python
# Splitting features and labels
X = breast_cancer_dataframe.drop(columns = 'label', axis = 1)
Y = breast_cancer_dataframe['label']
```

```python
print(X)
print(Y)
```

```python
# Splitting data into training and testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
# Standardizing the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Building the Neural Network
import tensorflow as tf
tf.random.set_seed(3) # tensorflow's random function will produce same values everytime
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential([
    Flatten(input_shape = (30,)),# X is in the form if matrix and we will flatten these values in a single 1d array
    Dense(20, activation = 'relu'), # rectified linear unit
    Dense(2 , activation = 'sigmoid')
])
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
history = model.fit(X_train, Y_train, validation_split = 0.1, epochs = 50)
```

```python
# Visualizing the training accuracy and validation accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc = 'upper left')
```

```python
# Visualizing the training loss and validation loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc = 'upper left')
```

```python
# Accuracy of model on test data
loss , accuracy = model.evaluate(X_test, Y_test)
```
